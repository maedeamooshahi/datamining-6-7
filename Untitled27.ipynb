```python
import numpy as np
import tensorflow as tf
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Conv1D, Dropout, Dense, Flatten, LSTM, MaxPooling1D, Bidirectional
from keras import backend as K
```

```python
train=pd.read_csv('/content/train.csv')
train
```

```python
train.drop(['Column1','label'], inplace=True,axis=1)
train
```

```python
dev=pd.read_csv('/content/dev.csv')
dev
```

```python
dev.drop(['Column1','label'], inplace=True,axis=1)
dev
```

```python
train=train.drop_duplicates(subset=None, keep='first', inplace=False)
train
```

```python
tokenizer = Tokenizer(num_words = 100000, oov_token='<oov>')
tokenizer.fit_on_texts(train['comment'])
```

```python
train_comments=tokenizer.texts_to_sequences(train['comment'])
test_comments=tokenizer.texts_to_sequences(dev['comment'])

training_padded=pad_sequences(train_comments,maxlen=100,truncating='post')
testing_padded=pad_sequences(test_comments,maxlen=100,truncating='post')
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, Dropout, Dense
covnet = tf.keras.Sequential([
                    tf.keras.layers.Embedding(100000,    32, input_length=100)
                    ])
covnet.summary()
```

```python
covnet.add(Conv1D(32, kernel_size=3, activation='elu', padding='same'))
covnet.add(tf.keras.layers.Flatten())
covnet.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
covnet.summary()
```

```python
covnet.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
prediction =covnet.fit(training_padded, train['label_id'], epochs=5, validation_data=(testing_padded, dev['label_id']))
```

```python
plt(prediction, 'accuracy')
```

```python
plt(prediction, 'loss')
```
