```python
Data['Heating_Load'] = np.where((Data.Heating_Load < 18), 0, Data.Heating_Load)
Data['Heating_Load'] = np.where((Data.Heating_Load >= 18), 1, Data.Heating_Load)
```

```python
from sklearn.ensemble import RandomForestClassifier
RandomFclassifier = RandomForestClassifier()
x = Data.drop(['Heating_Load','Cooling_Load'] , axis = 1)
y1 = Data['Heating_Load']
RanForclassifie = RandomFclassifier.fit(x, list(y1))
```
